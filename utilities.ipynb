```python
def get_stocks_with_industry(industry=None):
    output = None
    if None != industry:
        e_learning_stocks = list()
        for i, sym in enumerate(stocksdf["Symbol "].to_list()):
            try:
                if industry == stock_dictionary.get(sym).get('info').get('industry'):
                    e_learning_stocks.append(sym)
            except Exception as e:
                print(f"{i}:{e}")
        output = e_learning_stocks.copy()
    return output
```
